```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import os
```

```python
sp_iters_fp = r"data\iterations\SP"
mpg_iters_fp = r"data\iterations\MPG"
mpc_iters_fp = r"data\iterations\MPC"

sp_iters_dict = {}
mpg_iters_dict = {}
mpc_iters_dict = {}

for file in os.listdir(sp_iters_fp):
    sp_iters_txt_name = os.fsdecode(file)
    with open(f'{sp_iters_fp}\\{file}','r') as sp_txt:
        sp_txt_data = sp_txt.read()
        sp_iters_str = sp_txt_data.split(',')
        sp_iters_l = list(sp_iters_str)
        sp_iters_l = [int(n) for n in sp_iters_l if n!='']
        index = int( sp_iters_txt_name.replace('SP_iters_','').replace('.txt','') )
        sp_iters_dict[index] = sp_iters_l
    
for file in os.listdir(mpg_iters_fp):
    mpg_iters_txt_name = os.fsdecode(file)
    with open(f'{mpg_iters_fp}\\{file}','r') as mpg_txt:
        mpg_txt_data = mpg_txt.read()
        mpg_iters_str = mpg_txt_data.split(',')
        mpg_iters_l = list(mpg_iters_str)
        mpg_iters_l = [int(n) for n in mpg_iters_l if n!='']
        index = int( mpg_iters_txt_name.replace('MPG_iters_','').replace('.txt','') )
        mpg_iters_dict[index] = mpg_iters_l

for file in os.listdir(mpc_iters_fp):
    mpc_iters_txt_name = os.fsdecode(file)
    with open(f'{mpc_iters_fp}\\{file}','r') as mpc_txt:
        mpc_txt_data = mpc_txt.read()
        mpc_iters_str = mpc_txt_data.split(',')
        mpc_iters_l = list(mpc_iters_str)
        mpc_iters_l = [int(n) for n in mpc_iters_l if n!='']
        index = int( mpc_iters_txt_name.replace('MPC_iters_','').replace('.txt','') )
        mpc_iters_dict[index] = mpc_iters_l
```

```python
tot_n = len(mpc_iters_dict[0])
print(f'Simulation has been run {tot_n} times.')
```

```python
sp_iters_mean_dict = {key: np.mean(val) for key,val in sp_iters_dict.items()}
mpg_iters_mean_dict = {key: np.mean(val) for key,val in mpg_iters_dict.items()}
mpc_iters_mean_dict = {key: np.mean(val) for key,val in mpc_iters_dict.items()} 
```

```python
sp_iters_mean_dict_sorted = dict(sorted(sp_iters_mean_dict.items(), key=lambda item: item[1], reverse=True))
mpg_iters_mean_dict_sorted = {key:mpg_iters_mean_dict[key] for key in sp_iters_mean_dict_sorted.keys()} 
mpc_iters_mean_dict_sorted = {key:mpc_iters_mean_dict[key] for key in sp_iters_mean_dict_sorted.keys()} 

plt.figure(figsize=(20, 4))
plt.bar([str(n) for n in list(sp_iters_mean_dict_sorted.keys())], list(sp_iters_mean_dict_sorted.values()), label='SP', alpha=0.5, color='b')
plt.bar([str(n) for n in list(mpg_iters_mean_dict_sorted.keys())], list(mpg_iters_mean_dict_sorted.values()), label='MPG', alpha=0.5, color='r')
plt.bar([str(n) for n in list(mpc_iters_mean_dict_sorted.keys())], list(mpc_iters_mean_dict_sorted.values()), label='MPC', alpha=0.5, color='g')
plt.ylabel('Mean # iterations')
plt.xlabel('Index of topology')
plt.tick_params(axis='x', labelsize=8)
plt.title(f'Mean # iterations until success for various protocols (sorted wrt. SP) [sim. $\\times${tot_n}]')
plt.legend()
plt.margins(x=0.005, tight=True)
plt.show()
```

```python
sp_mpc_diff_dict = {key:mpc_iters_mean_dict[key]-sp_iters_mean_dict[key] for key in sp_iters_mean_dict.keys()}
sp_mpc_diff_dict_sorted = dict(sorted(sp_mpc_diff_dict.items(), key=lambda item: item[1], reverse=True))

plt.figure(figsize=(20, 4))
plt.bar([str(n) for n in list(sp_mpc_diff_dict_sorted.keys())], list(sp_mpc_diff_dict_sorted.values()),color=['red' if value >= 0 else 'green' for value in list(sp_mpc_diff_dict_sorted.values())])
plt.ylabel('Mean # iterations')
plt.xlabel('Index of topology')
plt.tick_params(axis='x', labelsize=8)
plt.title(f'Difference in mean # iterations between SP and MPC protocols [sim. $\\times${tot_n}]')
plt.margins(x=0.005, tight=True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axhline(5, color='black', linewidth=0.25)
plt.axhline(-5, color='black', linewidth=0.25)

red_patch = Patch(color='red', label='SP faster than MPC')
green_patch = Patch(color='green', label='MPC faster than SP')
plt.legend(handles=[green_patch,red_patch])
plt.show()
```
