```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
G = nx.Graph()
G.add_nodes_from(range(11))
G.add_edges_from([(0,1), (0,2), (2,3), (2,4), (3,6), (4,6), (1,5), (5,8), (6,8), (5,7), (7,9), (8,9)])

x = 1
pos = {
    0: (0,0),
    1: (1*x,0.5*x), 
    2: (1*x,-0.5*x), 
    3: (2*x,0), 
    4: (2*x,-1*x), 
    5: (3*x,0.5*x), 
    6: (3*x,-0.5*x), 
    7: (4*x,1*x), 
    8: (4*x,0), 
    9: (5*x,0.5*x),
    10: (7,3)
}

freq_dict = {
    0: 1,
    1: 0.4, 
    2: 0.6, 
    3: 0.1, 
    4: 0.5, 
    5: 0.4, 
    6: 0.6, 
    7: 0.2, 
    8: 0.8, 
    9: 1,
    10: 'temp'
}

color_map = []
for node in G.nodes():
        if node in [0,9]:
            color_map.append(matplotlib.colors.to_rgba('yellowgreen'))
        elif type(freq_dict[node]) == str:
             color_map.append(matplotlib.colors.to_rgba('white'))
        else:
            color_map.append((1 - freq_dict[node], 1 - freq_dict[node], 1, 1))

plt.figure(figsize=(6, 6))
nx.draw(G, pos, labels=freq_dict, node_size=1000, node_color=color_map, font_size=15, edgecolors='black')

plt.show()
```
